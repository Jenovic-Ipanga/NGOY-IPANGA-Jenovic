## Importation des Lib

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

## Chargement du datasets 

```python
data = pd.read_csv("data/studentscores.csv")
data
```

```python
data.head()
```

## Ananlyse des données 

```python
data.isnull()
```

```python
data.isnull().sum()
```

## 4 La division des donneedonnées 

```python
data.head()
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
X
```

```python
y
```

## 4.1 Séparation des donnees Train and Test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)
```

```python
print("X_train shape  : ", X_train.shape)
print("X_test shape   : ", X_test.shape)
print("y_train shape  : ", X_train.shape)
print("y_test shape   : ", X_test.shape)
```

## 5 Construire le Modèle 

```python
# Importation de l'algorithme de la Régression Linéaire
from sklearn.linear_model import LinearRegression
```

```python
model_lineaire = LinearRegression()
model_lineaire.fit(X_train, y_train)
```

## Faire la prediction

```python
y_score_predicts = model_lineaire.predict(X_test)
```

```python
y_score_predicts
```

```python
y_test
```

```python
resultat = pd.DataFrame({"y_test": y_test, "y_score_predicts": y_score_predicts})
resultat
```

## Évaluation du model 

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("MSE: ", mean_squared_error(y_test, y_score_predicts))
print("r2_score: ", r2_score(y_test, y_score_predicts))
```

## 6 La visualisation de Résultat 

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_train, y_train, color="blue", label="train data")
#plt.plot(X_test, y_score_predicts, color="red", label="ligne de la regression linéaire")
plt.plot(X_train, model_lineaire.predict(X_train), color="red", label="ligne de la regression linéaire")
plt.xlabel("les heures")
plt.ylabel("Scores")
plt.title("La régression linéaire")
plt.legend()
plt.show
```
